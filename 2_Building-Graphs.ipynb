#### 1. Creating a graph via declarative syntax.

The first way to create a graph is to explicitly add nodes and edges. 

````python
# Instantiate an undirected graph
g = nx.Graph()

# Add member nodes
for member_id in example.index:
    g.add_node(member_id)
    
# Add event node
g.add_node(233228364)

# Add edges
for member_id in example.index:
    g.add_edge(member_id, 233228364)
````
